```python
#import important libraires
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the first and second CSV file
data1 = pd.read_csv("/Users/USER/Desktop/WEMC WORK/sector CH4 agric CEDS.csv")
data2 = pd.read_csv("/Users/USER/Desktop/WEMC WORK/sector CH4 agric EDGAR.csv")

# Define the years range
years = [str(year) for year in range(2000, 2020 + 1)]

# Sum the emissions for the sector across all years for both datasets
total_emissions1 = data1[years].sum()
total_emissions2 = data2[years].sum()

# Plotting the aggregated data
plt.figure(figsize=(12, 8))

# Plot the first dataset
plt.plot(years, total_emissions1, color='b', label='CEDS')

# Plot the second dataset
plt.plot(years, total_emissions2, color='r', label='EDGAR')
#plt.ylim(40000, 52000)

# Adding titles and labels
plt.title('Total CH4 Emissions for CEDS and EDGAR for Agriculture(2000-2020)')
plt.xlabel('Year')
plt.ylabel('Total Emissions (kt)')
plt.xticks(rotation=45)
plt.legend()
#plt.grid(True)

# Display the plot
plt.tight_layout()
plt.show()
#saving to my pc
save_path = "/Users/USER/Desktop/WEMC WORK/total_emissions_comparison.png"
plt.savefig(save_path)
```
